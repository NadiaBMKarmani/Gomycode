```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
import seaborn as sb
```

```python
data = pd.read_csv("CC GENERAL.csv")
data.tail()
```

```python
#Preprocessing our data
def preprocess_data(new_data):
    new_data=new_data.drop("CUST_ID", axis=1)
    new_data['MINIMUM_PAYMENTS'].fillna(new_data['MINIMUM_PAYMENTS'].mean(),inplace=True) 
    new_data.dropna(axis=0, how='any', thresh=None,inplace=True)
    return new_data
data=preprocess_data(data)
data.head()
```

```python
data.isnull().sum()
```

```python
# 2. Use hierarchical clustering to identify the inherent groupings within your data.

from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(data)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
# 3. Plot the clusters. 
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data ['PURCHASES'] , data ["PAYMENTS"] , c= agglomerative[0], s=40)
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("PURCHASES")
ax.set_ylabel("PAYMENTS")
plt.colorbar(scatter)
```

```python
# 4. Plot the dendrogram.  
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(data, method="complete"))
```

```python
# 4.  Use K-means clustering. 
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=3, random_state=0)  #Cluster our data by choosing 3 as number of clusters
kmeans.fit(data)
labels=pd.DataFrame(kmeans.labels_)
labels
data['label']=kmeans.labels_
```

```python
kmeans.predict(data)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
# 5. Try different K values and select the best one. 

def calculate_sse(data, kmax):
  sse = []
  for k in kmax:
    kmeans = KMeans(n_clusters = k).fit(data)
    sse.append(kmeans.inertia_)
  return sse

# Elbow method

k=range(1,15)
sum_of_squared_distances=calculate_sse(data, k)

plt.plot(k, sum_of_squared_distances, 'bx-')
plt.title('Elbow method for optimal K')
plt.xlabel('Number of clusters K')
plt.ylabel('Sum of squared distances')
plt.show()
```

```python
# 6. Plot the clusters.

plt.scatter(data["PURCHASES"][data.label == 0],                   
            data["PAYMENTS"][data.label  == 0],s=40,c='magenta')
plt.scatter(data["PURCHASES"][data.label  == 1],          
            data["PAYMENTS"][data.label  == 1],s=40,c='yellow')
plt.scatter(data["PURCHASES"][data.label  == 2],          
            data["PAYMENTS"][data.label  == 2],s=40,c='green')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('PURCHASES')
plt.ylabel('PAYMENTS')
plt.show()
```

using Kmeans we get 3 clusters


```python
# Bonus: Search for another validation metric (Silhouette coefficient)

# Calculate Silhoutte Score
#
from sklearn.metrics import silhouette_score
score = silhouette_score(data, kmeans.labels_, metric='euclidean')
#
# Print the score
#
print('Silhouetter Score: %.3f' % score)
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

def silouette(data, kmax):  
  fig, ax = plt.subplots(2, 2, figsize=(15,8))
  for k in kmax:
    km = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=100, random_state=42)
    q, mod = divmod(k, 2)
    visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax[q-1][mod])
    visualizer.fit(data)

# Silouette method

k=range(2,5)
silouette(data, k)
```
